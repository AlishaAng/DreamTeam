```python
import pandas as pd
```

```python
#open file with image links
#contains links for all the inhibitors

df_image = pd.read_csv("inhibitors_images.csv")

#255 rows
df_image.head()
```

```python
#open file of wanted inhibitors
inhibitor_file = pd.read_csv("incomplete_inhibitors.csv")

#105 rows
inhibitor_file.head()
```

```python
#match inhibitor column from two files and extract image link(Picture column)
matching = inhibitor_file.join(df_image.set_index('Inhibitor'), on='Inhibitor')
new_db= matching[matching['Images'].notnull()]

#105 rows
new_db.head()
```

```python
#df into a csv file

new_db.to_csv('Final_inhibitors.csv', header=True, index=False)
```
