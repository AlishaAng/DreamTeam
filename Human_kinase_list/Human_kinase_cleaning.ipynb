The human_kinase_list.csv has not been tidied yet. The entry_name consists of both the entry_name and the number. This notebook contains the codes to clean them up. 

# Clean up the human_kinase_list

```python
#load kinase list as a database
human_kinase_db = pd.read_csv("/homes/zho30/DreamTeam/Human_kinase_list.csv")
human_kinase_db.head()
```

```python
#split up the entry name
human_kinase_db[['Uniprot_identifier','Uniprot_number']] = human_kinase_db['Entry name'].str.split("(",expand=True)
human_kinase_db.head()
```

```python
#clean up the database to remove the original column
human_kinase_db=human_kinase_db.drop(columns=['Entry name'], axis=1)
human_kinase_db.head()
```

```python
# tidy up the Uniprot_number
human_kinase_db['Uniprot_number']=human_kinase_db['Uniprot_number'].str.replace(")","")
human_kinase_db.head()
```

```python
#output the dataframe into a new csv
human_kinase_db.to_csv('clean_human_kinase.csv', sep=',', header=True)
```
